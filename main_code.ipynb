## Plotting a Graph of View Count vs Total Likes of some Youtube Trending Videos

```python
# Importing Libraries:
# Pandas - for Data Cleaning.
# Matplotlib - for Data Visulization.

import pandas as pd
from matplotlib import pyplot as plt

#Using a specific style(named "seaborn") for the graph visualization.
plt.style.use('seaborn')

# Reading the data from a csv file.
data = pd.read_csv('Youtube_Data_Sheet.csv')

# Separating the necessary columns.
view_count = data['view_count']    # Count of views in each video. It is a Pandas Series.
like_count = data['likes']         # Count of likes in each video. It is a Pandas Series.
ratio = data['ratio']              # This is Like to Dislike ratio. It is a Pandas Series.

# Predefining a Size for the plot to show.
plt.figure(figsize= (10, 7))

# Now plotting the Scatter Plot.
plt.scatter(view_count, like_count, c=ratio, cmap='summer', edgecolor='black', linewidth=1.2, alpha=0.65)

# Plotting a Colorbar and Setting a label for the colorbar for better reference in terms of visualization.
plt.colorbar().set_label('Like to Dislike Ratio')

# Setting up the log scale for X and Y axis:
plt.xscale('log')
plt.yscale('log')

# Setting up Titles and Labels for more Informations.
plt.title('Trending YouTube Videos', fontsize=25)
plt.xlabel('View Count', fontsize=15)
plt.ylabel('Total Likes', fontsize=15)

# tight_layout command for a better fit in small displays.
plt.tight_layout()

# Saving a .png Image file of the plot.
plt.savefig('Plotted Image')

# Showing the Plot on the screen.
plt.show()
```
